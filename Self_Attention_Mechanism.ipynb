```python
import numpy as np

# -----------------------------------
# Input Sequence
# -----------------------------------
# Example: 3 tokens with embedding size 4

X = np.array([
    [1, 0, 1, 0],
    [0, 2, 0, 2],
    [1, 1, 1, 1]
])

print("Input Matrix:\n")
print(X)

# -----------------------------------
# Initialize Weight Matrices
# -----------------------------------

np.random.seed(42)

W_Q = np.random.rand(4, 4)

W_K = np.random.rand(4, 4)

W_V = np.random.rand(4, 4)

# -----------------------------------
# Compute Queries, Keys, Values
# -----------------------------------

Q = np.dot(X, W_Q)

K = np.dot(X, W_K)

V = np.dot(X, W_V)

print("\nQuery Matrix:\n")
print(Q)

print("\nKey Matrix:\n")
print(K)

print("\nValue Matrix:\n")
print(V)

# -----------------------------------
# Attention Scores
# -----------------------------------

d_k = K.shape[-1]

scores = np.dot(Q, K.T) / np.sqrt(d_k)

print("\nAttention Scores:\n")
print(scores)

# -----------------------------------
# Softmax Function
# -----------------------------------

def softmax(x):

    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))

    return exp_x / np.sum(
        exp_x,
        axis=-1,
        keepdims=True
    )

attention_weights = softmax(scores)

print("\nAttention Weights:\n")
print(attention_weights)

# -----------------------------------
# Final Attention Output
# -----------------------------------

output = np.dot(attention_weights, V)

print("\nSelf-Attention Output:\n")
print(output)
```
